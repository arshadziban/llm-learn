```python
import torch
import torch.nn as nn
import pandas as pd
```

```python
tokens = ["the", "bank", "approved", "the", "loan"]
seq_len = len(tokens)
embedding_dim = 8
```

```python
embeddings = torch.rand(1, seq_len, embedding_dim)
```

```python
mask = torch.triu(torch.ones(seq_len, seq_len), diagonal=1)
mask = mask.masked_fill(mask == 1, float('-inf'))
```

```python
mha = nn.MultiheadAttention(
    embed_dim=embedding_dim,
    num_heads=2,
    batch_first=True
)
```

```python
output, attn_weights = mha(
    embeddings,
    embeddings,
    embeddings,
    attn_mask=mask
)
```

```python
attention_df = pd.DataFrame(
    attn_weights[0].detach().numpy(),
    index=tokens,
    columns=tokens
)

attention_df
```

Masked attention ensures that a decoder can only attend to past tokens, preventing it from seeing future words during text generation.
