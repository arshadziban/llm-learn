# Encoder Example

```python
import torch
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
encoder_model = AutoModel.from_pretrained("bert-base-uncased")
```

```python
sentence = "The bank approved the loan"

inputs = tokenizer(sentence, return_tensors="pt")
outputs = encoder_model(**inputs)
```

```python
print(outputs.last_hidden_state.shape)
```

### Meaning

- **1 sentence**
- **7 tokens** (after subword tokenization)
- Each token â†’ a **768-dimensional, context-aware embedding**

### What the Encoder Did

- Read the **entire sentence at once**
- Applied **self-attention**
- Created **context-aware representations**
- **Did not generate text**


# Decoder Example

```python
from transformers import AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("gpt2")
decoder_model = AutoModelForCausalLM.from_pretrained("gpt2")
```

```python
prompt = "The bank approved"

inputs = tokenizer(prompt, return_tensors="pt")
```

```python
output_ids = decoder_model.generate(
    **inputs,
    max_length=15,
    do_sample=False
)

generated_text = tokenizer.decode(output_ids[0], skip_special_tokens=True)
print(generated_text)
```
