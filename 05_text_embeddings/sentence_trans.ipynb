```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
query = "I like AI"
```

```python
sentences = [
    "I enjoy artificial intelligence",
    "Today is very hot"
]
```

```python
query_embedding = model.encode(query)
```

```python
sentence_embeddings = model.encode(sentences)
```

```python
similarities = cosine_similarity([query_embedding], sentence_embeddings)
```

```python
similarities
```

```python
for sentence, score in zip(sentences, similarities[0]):
    print(sentence, "→ similarity:", round(score, 2))
```

## Sentence Embedding Pipeline (Simple Flow)

Text  
↓  
**Sentence Transformer**  
↓  
**Embeddings** (numeric vectors)  
↓  
**Cosine Similarity**  
↓    
**Meaning comparison**
