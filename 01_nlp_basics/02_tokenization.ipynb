## Tokenization (Word and Subword)

Tokenization is the process of breaking text into smaller units called tokens so that a computer can process language.
Since models cannot understand raw text, tokenization is the first and most important step in NLP and LLMs.

**1. Word Tokenization**

Word tokenization splits a sentence into individual words, usually using spaces and punctuation.

**2. Subword Tokenization (Used in LLMs)**

Subword tokenization breaks words into smaller meaningful units instead of full words. This is the standard approach used by modern LLMs such as GPT, BERT, and T5.

# Word Tokenization

```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize

sentence = "I love natural language processing."
tokens = word_tokenize(sentence)

print(tokens)
```
